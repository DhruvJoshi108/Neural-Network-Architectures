```python
# 1.Write a Python program to create a tensor of shape (3, 3) filled with random numbers between 0 and 1.
#  Perform basic tensor operations including matrix addition, multiplication, and transpose.
#  Convert the resulting tensor into a NumPy array and then reshape it into a (1, 9) tensor. 
```

```python
import numpy as np
A = np.random.rand(3, 3)
print("Original Tensor (A):\n", A)
```

```python
B = A + A
print("Matrix Addition (A + A):\n", B)
```

```python
C = A @ A
print("Matrix Multiplication (A @ A):\n", C)
```

```python
D = C.T
print("Transpose of Matrix (C.T):\n", D)
```

```python
E= D.reshape(1, 9)
print("Reshaped Tensor (1, 9):\n", E)
```
